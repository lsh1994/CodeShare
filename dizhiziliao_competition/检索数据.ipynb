```python
wd=df['KEYWORDS']
wdlist=list(wd)
wdlist[:5]
```

```python
def wordcount(ary):
    count_dict = {}
    # 如果字典里有该单词则加1，否则添加入字典
    for str in ary:
        if str in count_dict.keys():
            count_dict[str]+= 1
        else:
            count_dict[str] = 1
    return count_dict
```

```python
wc=wordcount(wdlist)#关键词，数量
count_list=sorted(wc.items(),key=lambda x:x[1],reverse=True)# 降序排列
count_list[:30]
```

```python
ciyun=count_list[:1000]
name =[i[0] for i in ciyun]
value = [i[1] for i in ciyun]
```

```python
wordcloud = pyecharts.WordCloud(width=1000, height=620)
wordcloud.add("", name, value)
wordcloud
```
