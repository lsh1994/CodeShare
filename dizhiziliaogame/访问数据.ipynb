```python
df['TIME'] = pd.to_datetime(df['TIME'])
```

```python
df=df.set_index("TIME")
df.head()
```

```python
df.index
```

```python
df["2017-10"]
```

```python
timec=df.resample('m').count().iloc[:,0]
timec
```

```python
attr = list(timec.index.astype(str))
attr
```

```python
v1=list(timec.values)
v1
```

```python
import pyecharts
bar = pyecharts.Bar("访问数据按月份柱状图")
bar.add("", attr, v1, is_stack=True)
bar
```

```python
timec2=df.resample('w').count().iloc[:,0]
attr2 = list(timec2.index.astype(str))
v2=list(timec2.values)
line = pyecharts.Line("访问数据按星期折线图")
line.add("", attr2, v2, mark_point=["max", "average"])
line
```
