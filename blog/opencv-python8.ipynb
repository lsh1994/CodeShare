### 使用HSV对象跟踪

参考链接：[色彩空间中的HSL、HSV、HSB有什么区别？](https://www.zhihu.com/question/22077462)

```python
import numpy as np
```

```python
img=cv2.imread('e:/test.jpg')
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])
```

```python
mask = cv2.inRange(hsv, lower_blue, upper_blue)# 位于阈值之间的返回255，否则返回0
```

```python
res = cv2.bitwise_and(img,img, mask= mask)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(mask,cmap='gray')#注意白色部分（255）
```

```python
plt.imshow(cv2.cvtColor(res,cv2.COLOR_BGR2RGB))
```

上图表面上字体是黑色的，其实是因为被镂空成黑色背景（0,0,0）的缘故。

下面说明hsv与其他颜色的转换。

```python
# lower_blue = np.array([110,50,50])
# upper_blue = np.array([130,255,255])
green_bgr = np.uint8([[[0,255,0 ]]])
green_hsv=cv2.cvtColor(green_bgr,cv2.COLOR_BGR2HSV)
green_bgr
```

```python
lower_blue_hsv = np.uint8([[[110,50,50]]])
lower_blue_rgb=cv2.cvtColor(lower_blue_hsv,cv2.COLOR_HSV2RGB)
lower_blue_rgb
```

```python
upper_blue_hsv= np.uint8([[[130,255,255]]])
upper_blue_rgb=cv2.cvtColor(upper_blue_hsv,cv2.COLOR_HSV2BGR_FULL)
upper_blue_rgb
```

也可借助如图像编辑工具分析这些值。
