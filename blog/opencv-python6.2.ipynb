```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('e:/rotman.jpg')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)) #显示原图
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #将BGR图转为灰度图
plt.imshow(img,cmap='gray')
```

```python
ret,thr=cv2.threshold(img,60,255,cv2.THRESH_BINARY) # ret为阈值，thr为新的图像
print(ret,thr)
```

```python
plt.imshow(thr,cmap='gray')
```

```python
thr_inv=cv2.bitwise_not(thr) #按位，非
plt.imshow(thr_inv,cmap='gray')
```

```python
img = cv2.bitwise_or(img,img,mask=thr_inv)
plt.imshow(img,cmap='gray')
```
