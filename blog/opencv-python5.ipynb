```python
b,g,r = cv2.split(img) #splitnumpy
img = cv2.merge((r,g,b))
plt.imshow(img)
```

```python
b  =  img [:,:,0 ] #
print(b.shape)
plt.imshow(b)
```

```python
img1 = img

replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)#  aaaaaa | abcdefgh | hhhhhhh
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)# fedcba | abcdefgh | hgfedcb
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)#  gfedcb | abcdefgh | gfedcba
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)#  cdefgh | abcdefgh | abcdefg
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=[255,0,0])#

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

plt.show()
```
