```python
t1=cv2.getTickCount() #返回操作系统启动到现在毫秒时间
t1
```

```python
t2=cv2.getTickCount() #返回操作系统启动到现在毫秒时间
t2
```

```python
t=(t2-t1)/cv2.getTickFrequency()
t
```

### 计时工具%timeit指令

```python
x = 5
```

```python
%timeit y=x**2 # 使用 %timeit 多次运行某条命令从而进行比较
```

```python
%timeit y=x*x
```

```python
import numpy as np
%timeit y=np.square(x)
```

### useOptimized?

```python
cv2.useOptimized()#是否支持opencv的如sse2、avx优化代码
```

```python
img=cv2.imread('E:\bizhi.jpg')
```

```python
cv2.setUseOptimized(True)
%timeit res = cv2.medianBlur(img,299)
```

```python
cv2.setUseOptimized(False)#禁用优化代码
```

```python
cv2.useOptimized()
```

```python
%timeit res = cv2.medianBlur(img,299)
```
