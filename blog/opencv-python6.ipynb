## 操作前准备

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('e:/rotman.jpg')
plt.imshow(img)
```

```python
img2=cv2.imread('e:/cat.jpg')
plt.imshow(img2)
```

```python
print(img.shape)
print(img2.shape)
```

```python
img=img[50:350,50:350]
img2=img2[0:300,50:350] # 用numpy的切片操作置两张图片大小为300*300

# b,g,r=cv2.split(img)
# img=cv2.merge((r,g,b))
# b,g,r=cv2.split(img2)
# img2=cv2.merge((r,g,b))#重新组合波段
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img2)
```

## 图像相加

```python
s=cv2.add(img,img2) # 通过cv的add函数，也可以用参数标量
plt.imshow(s)
```

```python
s2=img+img2 # numpy的矩阵操作，区别前者饱和操作（最大值255），后者超过255时取模
plt.imshow(s2)
```

```python
dst = cv2.addWeighted(img,0.4,img2,0.6,0) # 按比例混合，形如 d=a*0.4+img2*0.6+0
plt.imshow(dst)
```
